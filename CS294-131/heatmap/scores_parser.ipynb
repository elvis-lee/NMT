```python
filename = "scores_1.txt"
f = open(filename)
lines = f.readlines()
num_line = len(lines)
```

```python
print(lines)
```

```python
output_filename = 'scores_1_processed.txt'
fo = open(output_filename, 'w')
```

```python
for i in range(num_line):
    line = lines[i]
    line = line.replace("][",',')
    line = line.replace("[[[",'[')
    line = line.replace("]]]",'],')
    if i == 0:
        line = line.replace('[',"[[")
    if i == num_line-1:
        line = line.replace("],","]]")
    print(line)
    fo.write(line)
```
